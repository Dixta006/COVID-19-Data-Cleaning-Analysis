## **Manual Cleaning**

```python
import pandas as pd

# Load dataset
df = pd.read_csv("//content/covid_messy_dataset.csv")

# Check missing values before cleaning
print("Missing values before cleaning:")
print(df.isnull().sum())

# Remove all rows containing at least one missing value
df_manual = df.dropna()

# Check missing values after cleaning
print("\nMissing values after cleaning:")
print(df_manual.isnull().sum())

# Compare dataset size
print("\nOriginal Shape:", df.shape)
print("Cleaned Shape:", df_manual.shape)

# Save cleaned dataset
df_manual.to_csv("covid_manual_cleaned.csv", index=False)

print("\nCleaned dataset saved successfully!")
```

```python
# Check total missing values
print(df.isnull().sum().sum())

# Remove all rows containing any NaN
df_manual = df.dropna(how='any')

# Reset index
df_manual = df_manual.reset_index(drop=True)

# Show cleaned data
display(df_manual.head(10))
```

# mean cleaning

```python
df_mean = df.copy()

df_mean['Cases'] = df_mean['Cases'].fillna(df_mean['Cases'].mean())
df_mean['Deaths'] = df_mean['Deaths'].fillna(df_mean['Deaths'].mean())
df_mean['Vaccinated'] = df_mean['Vaccinated'].fillna(df_mean['Vaccinated'].mean())

display(df_mean)
```

## **mode cleaning**

```python
df_mode = df.copy()

df_mode['Cases'] = df_mode['Cases'].fillna(df_mode['Cases'].mode()[0])
df_mode['Deaths'] = df_mode['Deaths'].fillna(df_mode['Deaths'].mode()[0])
df_mode['Vaccinated'] = df_mode['Vaccinated'].fillna(df_mode['Vaccinated'].mode()[0])

display(df_mode)
```

## **median cleaning**

```python
df_median = df.copy()

df_median['Cases'] = df_median['Cases'].fillna(df_median['Cases'].median())
df_median['Deaths'] = df_median['Deaths'].fillna(df_median['Deaths'].median())
df_median['Vaccinated'] = df_median['Vaccinated'].fillna(df_median['Vaccinated'].median())

display(df_median)
```

## comparison

```python
print("Original Missing Values")
print(df.isnull().sum())

print("\nManual Cleaning")
print(df_manual.isnull().sum())

print("\nMean Cleaning")
print(df_mean.isnull().sum())

print("\nMedian Cleaning")
print(df_median.isnull().sum())

print("\nMode Cleaning")
print(df_mode.isnull().sum())
```

```python
df_manual.to_csv("covid_manual_cleaned.csv", index=False)
```

# **Automatic Cleaning**

```python
df_auto = df.copy()

df_auto['Cases'].fillna(df_auto['Cases'].mean(), inplace=True)

df_auto['Deaths'].fillna(df_auto['Deaths'].mean(), inplace=True)

df_auto['Vaccinated'].fillna(df_auto['Vaccinated'].mean(), inplace=True)

display(df_auto)
```
